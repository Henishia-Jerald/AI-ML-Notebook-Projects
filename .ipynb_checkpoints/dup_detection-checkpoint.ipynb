<h3>Detecting Duplicate or Original Images Using AI</h3>
Use deep features from a pre-trained model like:
ResNet, VGG, EfficientNet, or CLIP<br>
These models extract robust feature vectors that encode image content and semantics.

<h3>Tools Required:</h3>
1.  TensorFlow / Keras<br>
2. Pretrained CNN (e.g., ResNet50, VGG16, or MobileNet)<br>
3. Cosine similarity or Euclidean distance to compare images

<h3> Importing Libraries</h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input 

import tensorflow as tf
from keras.preprocessing import image
from sklearn.metrics.pairwise import cosine_similarity

# from sqlalchemy import create_engine 
from sqlalchemy import create_engine, Table, Column, Integer, Text, LargeBinary, MetaData, insert,text

from PIL import Image
import io
```

```python
host= 'localhost'
database = 'images'
user = 'postgres'
password = 'amma'
port = 5432 

conn = create_engine(f"postgresql://{user}:{password}@{host}:{port}/{database}")

metadata = MetaData()

images = Table(
    "images", metadata,
    Column("id", Integer, primary_key=True),
    Column("name", Text),
    Column("image_data", LargeBinary)
)

# Read binary data from image file
with open("reunion.png", "rb") as f:
    image_data = f.read()

# Insert data
with conn.connect() as conn:
    stmt = insert(images).values(name="reunion.png", image_data=image_data)
    conn.execute(stmt)
    conn.commit()

```

```python
image = Image.open(io.BytesIO(image_bytes))

plt.imshow(image)
plt.axis('off')  # Optional: hide axes
plt.title("Image from PostgreSQL")
plt.show()

# Convert to NumPy array for processing
image_array = np.array(image)
print("Image shape:", image_array.shape)
```

<h3> Input Features (Images) </h3>

```python
IMG_PATH1 = r"C:\Users\Admin\Desktop\reunion.png"        # real image
IMG_PATH2 = r"C:\Users\Admin\Desktop\reunion1.png"       # AI image?

THRESHOLD = 0.95
```

<h3>Load the models</h3>

```python
# model = vgg16.VGG16(weights="imagenet")
model = VGG16(weights='imagenet', include_top=False, pooling='avg')
```

<h3>Loading and Preprocessing the Images</h3>

```python
def extract_features(img_path):
    img = image.load_img(img_path, target_size=(224, 224))
    features = image.img_to_array(img)
    features = np.expand_dims(features, axis=0)
    features = preprocess_input(features)
    features = model.predict(features)
    return features.flatten()
```

<h3>Finding Cosine Similarity of the Images</h3>

```python
def get_cos_similarity(feature1,feature2):
    similarity = cosine_similarity([feature1], [feature2])[0][0]
    return similarity
```

<h3>Displaying Images</h3>

```python
def display_images(image_path1,image_path2,cosine_similarity):
    fig, axes = plt.subplots(1, 2, figsize=(10, 5))
    
    img1 = image.load_img(image_path1)
    img2 = image.load_img(image_path2)
    
    axes[0].imshow(img1)
    axes[0].axis('off')
    axes[0].set_title("Real Image")
    
    axes[1].imshow(img2)
    axes[1].axis('off')
    axes[1].set_title("Test Image")
    
    plt.suptitle(f"Cosine Similarity: {cosine_similarity:.4f}", fontsize=14)
    plt.tight_layout()
    plt.show()
    
    if similarity > THRESHOLD:
        print("Test image is possibly real as it is similar to real image")
    else:
        print("Test image is possibly AI-generated as it is likely different")
```

<h3> Calling all the functions </h3>

```python
feature1 = extract_features(IMG_PATH1)
feature2 = extract_features(IMG_PATH2)
similarity = get_cos_similarity(feature1,feature2)
display_images(IMG_PATH1,IMG_PATH2,similarity)
```
